```python
import plotly.express as px
import pandas as pd


df = pd.read_csv(r'C:\Users\sai kiran\OneDrive\Desktop\lung_cancer\dataset.csv')

column_to_compare = 'LUNG_CANCER'
categorical_columns = [col for col in df.columns if col !='AGE']
categorical_columns = [col for col in categorical_columns if col != column_to_compare]

for col in categorical_columns:
    fig = px.histogram(df, x=col, color=column_to_compare, barmode='group',
                       title=f'Comparison of "{column_to_compare}" with "{col}"',
                       labels={col: col, column_to_compare: column_to_compare})
    fig.update_layout(xaxis_title=col, yaxis_title='Count', bargap=0.1)
    fig.show()
```

```python
#for interactive heatmaps

import plotly.graph_objects as go

for col in categorical_columns:
    cross_tab = pd.crosstab(df[column_to_compare], df[col])

    fig = go.Figure(data=go.Heatmap(
        z=cross_tab.values,
        x=cross_tab.columns,
        y=cross_tab.index,
        colorscale='Blues',
        text=cross_tab.values,
        hoverinfo='text'
    ))

    fig.update_layout(
        title=f'Cross-tabulation of "{column_to_compare}" and "{col}"',
        xaxis_title=col,
        yaxis_title=column_to_compare
    )

    fig.show()
```
